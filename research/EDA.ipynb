# Exploratory data analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/okcupid_profiles.csv")
```

```python
# 1. Basic Summary Statistics
def basic_summary(data):
    print("Data Overview\n=============")
    print("Shape:", data.shape)
    print("Numerical features: ", list(data.select_dtypes(include=['int64', 'float64']).columns))
    print("Categorical features: ", list(data.select_dtypes(exclude=['int64', 'float64']).columns))
basic_summary(data)
```

```python
# 2. Visualizing Distributions of Numerical Features
def numerical_distributions(data):
    numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns

    # Plot distributions for numerical features
    plt.figure(figsize=(12, 8))
    for i, col in enumerate(numerical_cols, 1):
        plt.subplot(3, 3, i)
        sns.histplot(data[col], kde=True, bins=30, color='skyblue')
        plt.title(f'Distribution of {col}')
        plt.tight_layout()
    plt.show()

numerical_distributions(data)
```

```python
# 3. Visualizing Distributions of Categorical Features
def categorical_distributions(data):
    categorical_cols = ["status", "sex", "body_type", "drugs", "drinks", "orientation", "pets", "offspring", "smokes"]



       
    plt.figure(figsize=(12, 8))
    for i, col in enumerate(categorical_cols, 1):
        plt.subplot(3, 3, i)
        sns.countplot(data=data, x=col, color='skyblue')
        plt.title(f'{col}')
        plt.xlabel('')
        # Rotate the x-tick labels to prevent overlap
        plt.xticks(rotation=45, ha="right")
        # Ensure tight layout to avoid clipping
        plt.tight_layout()
    # Show all the plots
    plt.show()

categorical_distributions(data)
```

```python
# 4. Correlation Heatmap for Numerical Features
def correlation_heatmap(data):
    numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
    correlation_matrix = data[numerical_cols].corr()
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title('Correlation Heatmap of Numerical Features')
    plt.show()

correlation_heatmap(data)
```

```python
# 5. Pairplot for Numerical Features
def pairplot_numerical(data):
    numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
    sns.pairplot(data[numerical_cols], kind='scatter', plot_kws={'alpha':0.6, 's':50})
    plt.show()
    
pairplot_numerical(data)
```

```python
# 6. Analyzing Missing Data Patterns
def missing_data_pattern(data):
    plt.figure(figsize=(10, 8))
    sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
    plt.title('Missing Data Heatmap')
    plt.show()

missing_data_pattern(data)
```
